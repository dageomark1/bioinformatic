```python
import pandas as pd

df = pd.DataFrame({
    "Gender": ["M","F","M","F"],
    "Region": ["North","North","South","South"],
    "Sales": [100,200,150,300]
})
```

```python
df.pivot_table(
    values="Sales",
    index="Gender",
    aggfunc="sum"
)
```

```python
df.pivot_table(
    values="Sales",
    index="Gender",
    columns="Region",
    aggfunc="sum"
)
```
